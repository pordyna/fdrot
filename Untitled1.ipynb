```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(1, 51, dtype = 'float64')
y = np.arange(1, 18, dtype = 'float64')
```

```python
x = x[:,None]
x = x
```

```python
y = y[None,:]
```

```python
y
```

```python
f = np.exp(-2/10**2 * ((x - 26)**2 +  (y - 9)**2))
f[f.nonzero()]
```

```python
vmin = np.min(f[f.nonzero()])
vmax = np.max(f[f.nonzero()])
```

```python
fig, ax = plt.subplots(1,1,figsize=(20,20))
plt.imshow(f, cmap = 'summer' ,vmin=vmin ,vmax=vmax )
```
